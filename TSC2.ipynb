Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
````

````python
def normalize(X):

    X = np.divide(X, 255)
    return [np.expand_dims(rgb2gray(x), axis=2) for x in X]
````

````python
def normalize2(image_data):
    """
    Normalize the image data with Min-Max scaling to a range of [0.1, 0.9]
    :param image_data: The image data to be normalized
    :return: Normalized image data
    """
    return 0.1 + 0.8 * image_data / 255
````

````python
def preprocess(dataset, algorithm):
    after = []
    for e in dataset:
        after.append(algorithm(e))
    return np.array(after)
````

### Model Architecture

````python
### This is the same model fro LeNet Lab. This yields a validation accuracy of 91%
def LeNet5(x):    
    # Arguments used for tf.truncated_normal
    # randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    
    # Layer 1: Convolutional. 32x32x1 => 28x28x12
    L1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 12), mean=mu, stddev=sigma))
    L1_b = tf.Variable(tf.zeros(12))
    L1_conv = tf.nn.conv2d(x, L1_W, [1, 1, 1, 1], 'VALID') + L1_b
    
    # Activation.
    L1_act = tf.nn.relu(L1_conv)

    
    # Pooling. 28x28x12 => 14x14x32.
    L1_out = tf.nn.max_pool(L1_act, [1, 2, 2, 1], [1, 2, 2, 1], 'VALID')
    
    # Layer 2: Convolutional. 14x14x32 => 10x10x32.
    L2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 12, 32), mean=mu, stddev=sigma))
    L2_b = tf.Variable(tf.zeros(32))
    L2_conv = tf.nn.conv2d(L1_out, L2_W, [1, 1, 1, 1], 'VALID') + L2_b
    
    # Activation.
    L2_act = tf.nn.relu(L2_conv)
    

    
    # Pooling. 10x10x32 => 5x5x32.
    L2_pool = tf.nn.max_pool(L2_act, [1, 2, 2, 1], [1, 2, 2, 1], 'VALID')
    
    # Flatten. 5x5x32 => Output = 800.
    L2_out = tf.contrib.layers.flatten(L2_pool)
    
    # Layer 3: Fully Connected. 800 => 480.
    L3_weights = tf.Variable(tf.truncated_normal(shape=(800, 480), mean=mu, stddev=sigma))
    L3_biases = tf.Variable(tf.zeros(480))
    L3_linear = tf.matmul(L2_out, L3_weights) + L3_biases
    
    # Activation.
    L3_out = tf.nn.relu(L3_linear)
    
    L3_out = tf.nn.dropout(L3_out, keep_prob=0.85)
    
    # Layer 4: Fully Connected. 480 => 240.
    L4_weights = tf.Variable(tf.truncated_normal(shape=(480, 240), mean=mu, stddev=sigma))
    L4_biases = tf.Variable(tf.zeros(240))
    L4_linear = tf.matmul(L3_out, L4_weights) + L4_biases
    
    # Activation.
    L4_out = tf.nn.relu(L4_linear)
    
    L4_out = tf.nn.dropout(L4_out, keep_prob=0.85)
    
    # Layer 5: Fully Connected. 240 => 43.
    L5_weights = tf.Variable(tf.truncated_normal(shape=(240, 43), mean=mu, stddev=sigma))
    L5_biases = tf.Variable(tf.zeros(43))
    logits = tf.matmul(L4_out, L5_weights) + L5_biases
    
#     print(logits.shape)
    
    return logits
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
# Tuning parameters
EPOCHS = 100
BATCH_SIZE = 128
LR = 0.001
keep_prob = tf.placeholder(tf.float32)
````

````python
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)
````

````python
logits = LeNet5(x)
````

````python
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
# TODO try different optimizer
optimizer = tf.train.AdamOptimizer(learning_rate = LR)
#optimizer = tf.train.GradientDescentOptimizer(learning_rate = LR)
training_operation = optimizer.minimize(loss_operation)
````

````python
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
````

### Training and Validation

````python
saver = tf.train.Saver()

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.8})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
````
